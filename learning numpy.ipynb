## Calculus

Integrals and derivatives


````python
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)

# Derivatives
dydx = np.gradient(y, x)

# Integrals
y_int = np.cumsum(y) * (x[1] - x[0])
````

````python
plt.plot(x,y)
plt.plot(x, dydx)
plt.plot(x,y_int)
````

### Question 1

Let  $ y = e^{\frac{-x}{10}}\sin(x)$. Consider 10000x intervals in the range [0,10]

1. Plot the function y vs x in the range [0,10].
2. Compute the mean and standard deviation of y for x values in [4,7].
3. For x in the range [4,7], find the value y_m such that 80% of y values are less than y_m.
4. Plot dy/dx vs x
5. Find the locations where dy/dx = 0.

````python
#1
N = 10000
x = np.linspace(0,10, N + 1)
y = np.exp(-x/10) * np.sin(x)

plt.plot(x,y)
````

````python
# 2
range_value_y = y[(x>=4)*(x<=7)]
mean = np.mean(range_value_y)
print(f'The mean is: {mean}')

standard_dev = np.std(range_value_y)
print(f'The standard deviation is: {standard_dev}')
````

````python
# 3
percentile = np.percentile(range_value_y, 80)
print(f'The percentile is: {percentile}')
````

````python
# 4 
der_x = np.gradient(y,x)
plt.plot(x,der_x)
````

````python
# 5
# The points where dydx is zero which is the same as the root of the equation
d_root = x[1:][der_x[1:] * der_x[:-1] < 0]
print(f'The roots are: {d_root}')
````

### Question 3

Consider the flower petal $ r(\theta) = 1 + \frac{3}{4}\sin(3\theta)$  for $0<= \theta <= 2\pi $
1. Make a plot of the flower (find x and y points)
2. Compute the area using the calculus formula A = integral( 1/2 * r^2 d_teta) with an interval of 0 to 2 pi.
3. Compute the arclength using the calculus formula L = integral(sqr(r^2 + (dr/d_teta)^2 d_teta))

````python
# 1
theta = np.linspace(0, 2*np.pi, 1000)
r = 1 + 3/4 * np.sin(3*theta)
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y)
````

````python
 # 2
A = 1/2 * sum((r**2)) * (theta[1] - theta[0])
print(A)
````

````python
# 3
arc_length: float = sum(np.sqrt(r**2 + np.gradient(r, theta)**2)) * (theta[1] - theta[0])
print(arc_length)
````

### Question 4

The power emitted by a blackbody is $ P = (A \sigma \epsilon)T^4 $. After measuring the temperature of a star you find that T(t) = T_o * (1 / (1 + e^-kt)). Plot the total energy emitted by the star as a function of time using the fact that E(t) = integral(P(t')d_t') 

````python
kt = np.linspace(0, 3, 100)
P = (1 / (1 + np.exp(-kt)))**4
E = np.cumsum(P) * (kt[1] - kt[0])
````

````python
plt.plot(kt, E)
plt.xlabel('$kt$', fontsize=20)
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$', fontsize=20)
````
